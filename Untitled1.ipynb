```python
import simpy
import numpy as np

# -----------------------------
# Parameter
# -----------------------------
SIM_DAYS = 14
BEDS = 100
LAMBDA = 8                # Ø Ankünfte pro Tag
LOS_SHAPE = 2.0           # Semi-Markov: Formparameter
LOS_SCALE = 3.0           # Semi-Markov: Skalenparameter
SEED = 42

rng = np.random.default_rng(SEED)

# Statistik
daily_occupancy = []
max_occupancy = 0

# -----------------------------
# Patientenprozess (Entity)
# -----------------------------
def patient(env, hospital, name):
    global max_occupancy

    arrival_time = env.now

    with hospital.request() as req:
        yield req  # warten, bis Bett frei

        # Bett belegt
        current = hospital.count
        max_occupancy = max(max_occupancy, current)

        # Semi-Markov: Aufenthaltsdauer
        los = rng.gamma(LOS_SHAPE, LOS_SCALE)
        yield env.timeout(los)

        # Entlassung automatisch beim Verlassen des with-Blocks

# -----------------------------
# Ankunftsprozess (Poisson)
# -----------------------------
def arrival_process(env, hospital):
    i = 0
    while env.now < SIM_DAYS:
        # Exponentielle Zwischenankunftszeit
        interarrival = rng.exponential(1 / LAMBDA)
        yield env.timeout(interarrival)

        i += 1
        env.process(patient(env, hospital, f"Patient {i}"))

# -----------------------------
# Monitoring der Auslastung
# -----------------------------
def monitor(env, hospital):
    while True:
        daily_occupancy.append(hospital.count)
        yield env.timeout(1)

# -----------------------------
# Simulation starten
# -----------------------------
env = simpy.Environment()
hospital = simpy.Resource(env, capacity=BEDS)

env.process(arrival_process(env, hospital))
env.process(monitor(env, hospital))

env.run(until=SIM_DAYS)

# -----------------------------
# Ergebnisse
# -----------------------------
print(f"Simulationsdauer: {SIM_DAYS} Tage")
print(f"Bettenkapazität: {BEDS}")
print(f"Maximale Belegung: {max_occupancy}")
print(f"Durchschnittliche Belegung: {np.mean(daily_occupancy):.1f}")
```
